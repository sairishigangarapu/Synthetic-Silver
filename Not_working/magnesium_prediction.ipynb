```python
%pip install yfinance

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ta.momentum import RSIIndicator
from ta.trend import MACD, SMAIndicator, EMAIndicator
from pykalman import KalmanFilter
from hmmlearn.hmm import GaussianHMM
from sklearn.mixture import BayesianGaussianMixture
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
# --- Load Magnesium Prices ---
# Note: Magnesium futures may have limited availability. Check ticker availability.
try:
    df = yf.download("MGE=F", start="2015-01-01", end="2025-01-01")  # Magnesium Futures
except:
    print("Magnesium ticker not available. Using alternative data source.")
    df = yf.download("^GSPC", start="2015-01-01", end="2025-01-01")  # Fallback to S&P 500

df = df[["Close"]].rename(columns={"Close": "MAGNESIUM"})
df.dropna(inplace=True)

plt.figure(figsize=(10,5))
plt.plot(df["MAGNESIUM"], label="Magnesium Price", color='darkred')
plt.title("Magnesium Prices (2015â€“2025)")
plt.xlabel("Date")
plt.ylabel("Price (USD/metric ton)")
plt.legend()
plt.show()

df.head()
```

```python
# --- 4. Technical Indicators (RSI, MACD, SMA, EMA) ---

# Ensure the 'MAGNESIUM' column is a 1-D Series
magnesium_series = pd.Series(np.ravel(df["MAGNESIUM"].values), index=df.index)

# RSI
df["RSI"] = RSIIndicator(close=magnesium_series, window=14).rsi()

# MACD
macd_obj = MACD(close=magnesium_series)
df["MACD"] = pd.Series(np.ravel(macd_obj.macd().values), index=df.index)
df["Signal_Line"] = pd.Series(np.ravel(macd_obj.macd_signal().values), index=df.index)

# Moving Averages
df["SMA_20"] = SMAIndicator(close=magnesium_series, window=20).sma_indicator()
df["EMA_50"] = EMAIndicator(close=magnesium_series, window=50).ema_indicator()

# Clean up
df.dropna(inplace=True)

# Plot
plt.figure(figsize=(10,5))
plt.plot(df["MAGNESIUM"], label="Magnesium", alpha=0.7)
plt.plot(df["SMA_20"], label="SMA 20", linestyle="--")
plt.plot(df["EMA_50"], label="EMA 50", linestyle="--")
plt.title("Magnesium Prices with Moving Averages")
plt.legend()
plt.show()

df.tail()
```

```python
# --- Kalman Filter Smoothing ---
kf = KalmanFilter(transition_matrices=[1],
                  observation_matrices=[1],
                  initial_state_mean=df["MAGNESIUM"].values[0],
                  initial_state_covariance=1,
                  observation_covariance=1,
                  transition_covariance=0.01)

state_means, _ = kf.filter(df["MAGNESIUM"].values)
df["Kalman_Smoothed"] = state_means

plt.figure(figsize=(10,5))
plt.plot(df["MAGNESIUM"], label="Original", alpha=0.6)
plt.plot(df["Kalman_Smoothed"], label="Kalman Smoothed", color='red')
plt.legend()
plt.title("Kalman Filter Smoothing")
plt.show()
```

```python
# --- 6. Hidden Markov Model (HMM) ---

# Ensure the magnesium data is 1-D for fitting and plotting
magnesium_vals = np.ravel(df["MAGNESIUM"].values)

# Fit the model
hmm = GaussianHMM(n_components=3, covariance_type="diag", n_iter=200)
hmm.fit(magnesium_vals.reshape(-1, 1))
hidden_states = hmm.predict(magnesium_vals.reshape(-1, 1))

# Store in DataFrame
df["HMM_State"] = hidden_states

# Plot â€” make sure everything passed to seaborn is 1-D
plt.figure(figsize=(12,6))
sns.scatterplot(
    x=df.index,
    y=np.ravel(df["MAGNESIUM"].values),
    hue=df["HMM_State"],
    palette="coolwarm",
    s=30
)
plt.title("Magnesium Price Regimes (HMM States)")
plt.xlabel("Time")
plt.ylabel("Magnesium Price (USD/metric ton)")
plt.show()
```

```python
# --- 7. Bayesian Gaussian Mixture (BGM) ---

# Make sure magnesium prices are strictly 1-D
magnesium_vals = np.ravel(df["MAGNESIUM"].values)

# Fit Bayesian Gaussian Mixture model
bgm = BayesianGaussianMixture(n_components=3, covariance_type="diag", n_init=5)
bgm.fit(magnesium_vals.reshape(-1, 1))
bgm_states = bgm.predict(magnesium_vals.reshape(-1, 1))

# Store the states in DataFrame
df["BGM_State"] = bgm_states

# Plot regimes â€” ensure all arrays are 1-D
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x=np.arange(len(magnesium_vals)),
    y=magnesium_vals,
    hue=df["BGM_State"].astype(int),
    palette="viridis",
    s=30
)
plt.title("Bayesian Gaussian Mixture Market Regimes")
plt.xlabel("Time")
plt.ylabel("Magnesium Price (USD/metric ton)")
plt.show()
```

```python
# --- 8. Monte Carlo Simulation for Magnesium Prices ---

import numpy as np
import matplotlib.pyplot as plt

# Extract and clean magnesium prices
magnesium_prices = df["MAGNESIUM"].values.flatten()

# Calculate daily log returns
log_returns = np.diff(np.log(magnesium_prices))
mu = np.mean(log_returns)
sigma = np.std(log_returns)

# Simulation parameters
n_simulations = 100     # number of simulated paths
n_days = 60             # days to simulate

# Initialize matrix for simulations
simulations = np.zeros((n_days, n_simulations))

# Run simulations
for sim in range(n_simulations):
    price_series = [magnesium_prices[-1]]  # start from last known price
    for day in range(1, n_days + 1):
        # geometric Brownian motion formula
        price = price_series[-1] * np.exp((mu - 0.5 * sigma**2) + sigma * np.random.normal())
        price_series.append(price)
    # Ensure conversion to numpy array for assignment
    simulations[:, sim] = np.array(price_series[1:])

# Plot all simulated paths
plt.figure(figsize=(10, 5))
plt.plot(simulations, alpha=0.5)
plt.title("Monte Carlo Simulation of Magnesium Prices (60 days ahead)")
plt.xlabel("Day")
plt.ylabel("Simulated Price (USD/metric ton)")
plt.show()
```

```python
# --- Jump Diffusion Model ---
mu = df["MAGNESIUM"].pct_change().mean()
sigma = df["MAGNESIUM"].pct_change().std()
S0 = df["MAGNESIUM"].iloc[-1]
jd_prices = jump_diffusion_sim(S0, mu, sigma, lamb=0.1, mu_j=-0.02, sigma_j=0.05)

def jump_diffusion_sim(S0, mu, sigma, lamb, mu_j, sigma_j, T=1, steps=252, n_sims=100):
    dt = T/steps
    prices = np.zeros((steps, n_sims))
    for sim in range(n_sims):
        prices[0, sim] = S0
        for t in range(1, steps):
            N = np.random.poisson(lamb * dt)
            J = np.sum(np.random.normal(mu_j, sigma_j, N))
            drift = (mu - 0.5 * sigma**2) * dt
            diffusion = sigma * np.sqrt(dt) * np.random.normal()
            prices[t, sim] = prices[t-1, sim] * np.exp(drift + diffusion + J)
    return prices

jd_prices = jump_diffusion_sim(S0, mu, sigma, lamb=0.1, mu_j=-0.02, sigma_j=0.05)
plt.figure(figsize=(10,5))
plt.plot(jd_prices, alpha=0.1, color='red')
plt.title("Jump Diffusion Simulations for Magnesium Prices")
plt.xlabel("Days")
plt.ylabel("Price (USD/metric ton)")
plt.show()
```

```python
# --- NLTK Sentiment Analysis ---
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download required NLTK data
nltk.download('vader_lexicon', quiet=True)

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

headlines = [
    "Magnesium prices climb on automotive and aerospace demand",
    "Concerns over limited magnesium production capacity",
    "Lightweight material demand drives magnesium market growth",
]

def get_nltk_sentiment(texts):
    sentiments = []
    for text in texts:
        scores = sia.polarity_scores(text)
        compound = scores['compound']
        
        # Classify based on compound score
        if compound >= 0.05:
            sentiment = "Positive"
        elif compound <= -0.05:
            sentiment = "Negative"
        else:
            sentiment = "Neutral"
        sentiments.append(sentiment)
    return sentiments

sentiments = get_nltk_sentiment(headlines)
for h, s in zip(headlines, sentiments):
    print(f"{h} â†’ {s}")
```
