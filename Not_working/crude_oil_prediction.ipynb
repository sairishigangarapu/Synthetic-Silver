```python
%pip install yfinance

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ta.momentum import RSIIndicator
from ta.trend import MACD, SMAIndicator, EMAIndicator
from pykalman import KalmanFilter
from hmmlearn.hmm import GaussianHMM
from sklearn.mixture import BayesianGaussianMixture
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import warnings
warnings.filterwarnings("ignore")
```

```python
# --- Load Crude Oil Prices ---
df = yf.download("CL=F", start="2015-01-01", end="2025-01-01")  # WTI Crude Oil Futures
df = df[["Close"]].rename(columns={"Close": "CRUDE_OIL"})
df.dropna(inplace=True)

plt.figure(figsize=(10,5))
plt.plot(df["CRUDE_OIL"], label="Crude Oil Price", color='darkgreen')
plt.title("Crude Oil Prices (20152025)")
plt.xlabel("Date")
plt.ylabel("Price (USD/barrel)")
plt.legend()
plt.show()

df.head()
```

```python
# --- 4. Technical Indicators (RSI, MACD, SMA, EMA) ---

# Ensure the 'CRUDE_OIL' column is a 1-D Series
crude_oil_series = pd.Series(np.ravel(df["CRUDE_OIL"].values), index=df.index)

# RSI
df["RSI"] = RSIIndicator(close=crude_oil_series, window=14).rsi()

# MACD
macd_obj = MACD(close=crude_oil_series)
df["MACD"] = pd.Series(np.ravel(macd_obj.macd().values), index=df.index)
df["Signal_Line"] = pd.Series(np.ravel(macd_obj.macd_signal().values), index=df.index)

# Moving Averages
df["SMA_20"] = SMAIndicator(close=crude_oil_series, window=20).sma_indicator()
df["EMA_50"] = EMAIndicator(close=crude_oil_series, window=50).ema_indicator()

# Clean up
df.dropna(inplace=True)

# Plot
plt.figure(figsize=(10,5))
plt.plot(df["CRUDE_OIL"], label="Crude Oil", alpha=0.7)
plt.plot(df["SMA_20"], label="SMA 20", linestyle="--")
plt.plot(df["EMA_50"], label="EMA 50", linestyle="--")
plt.title("Crude Oil Prices with Moving Averages")
plt.legend()
plt.show()

df.tail()
```

```python
# --- Kalman Filter Smoothing ---
kf = KalmanFilter(transition_matrices=[1],
                  observation_matrices=[1],
                  initial_state_mean=df["CRUDE_OIL"].values[0],
                  initial_state_covariance=1,
                  observation_covariance=1,
                  transition_covariance=0.01)

state_means, _ = kf.filter(df["CRUDE_OIL"].values)
df["Kalman_Smoothed"] = state_means

plt.figure(figsize=(10,5))
plt.plot(df["CRUDE_OIL"], label="Original", alpha=0.6)
plt.plot(df["Kalman_Smoothed"], label="Kalman Smoothed", color='red')
plt.legend()
plt.title("Kalman Filter Smoothing")
plt.show()
```

```python
# --- 6. Hidden Markov Model (HMM) ---

# Ensure the crude oil data is 1-D for fitting and plotting
crude_oil_vals = np.ravel(df["CRUDE_OIL"].values)

# Fit the model
hmm = GaussianHMM(n_components=3, covariance_type="diag", n_iter=200)
hmm.fit(crude_oil_vals.reshape(-1, 1))
hidden_states = hmm.predict(crude_oil_vals.reshape(-1, 1))

# Store in DataFrame
df["HMM_State"] = hidden_states

# Plot  make sure everything passed to seaborn is 1-D
plt.figure(figsize=(12,6))
sns.scatterplot(
    x=df.index,
    y=np.ravel(df["CRUDE_OIL"].values),
    hue=df["HMM_State"],
    palette="coolwarm",
    s=30
)
plt.title("Crude Oil Price Regimes (HMM States)")
plt.xlabel("Time")
plt.ylabel("Crude Oil Price (USD/barrel)")
plt.show()
```

```python
# --- 7. Bayesian Gaussian Mixture (BGM) ---

# Make sure crude oil prices are strictly 1-D
crude_oil_vals = np.ravel(df["CRUDE_OIL"].values)

# Fit Bayesian Gaussian Mixture model
bgm = BayesianGaussianMixture(n_components=3, covariance_type="diag", n_init=5)
bgm.fit(crude_oil_vals.reshape(-1, 1))
bgm_states = bgm.predict(crude_oil_vals.reshape(-1, 1))

# Store the states in DataFrame
df["BGM_State"] = bgm_states

# Plot regimes  ensure all arrays are 1-D
plt.figure(figsize=(12, 6))
sns.scatterplot(
    x=np.arange(len(crude_oil_vals)),
    y=crude_oil_vals,
    hue=df["BGM_State"].astype(int),
    palette="viridis",
    s=30
)
plt.title("Bayesian Gaussian Mixture Market Regimes")
plt.xlabel("Time")
plt.ylabel("Crude Oil Price (USD/barrel)")
plt.show()
```

```python
# --- 8. Monte Carlo Simulation for Crude Oil Prices ---

import numpy as np
import matplotlib.pyplot as plt

# Extract and clean crude oil prices
crude_oil_prices = df["CRUDE_OIL"].values.flatten()

# Calculate daily log returns
log_returns = np.diff(np.log(crude_oil_prices))
mu = np.mean(log_returns)
sigma = np.std(log_returns)

# Simulation parameters
n_simulations = 100     # number of simulated paths
n_days = 60             # days to simulate

# Initialize matrix for simulations
simulations = np.zeros((n_days, n_simulations))

# Run simulations
for sim in range(n_simulations):
    price_series = [crude_oil_prices[-1]]  # start from last known price
    for day in range(1, n_days + 1):
        # geometric Brownian motion formula
        price = price_series[-1] * np.exp((mu - 0.5 * sigma**2) + sigma * np.random.normal())
        price_series.append(price)
    # Ensure conversion to numpy array for assignment
    simulations[:, sim] = np.array(price_series[1:])

# Plot all simulated paths
plt.figure(figsize=(10, 5))
plt.plot(simulations, alpha=0.5)
plt.title("Monte Carlo Simulation of Crude Oil Prices (60 days ahead)")
plt.xlabel("Day")
plt.ylabel("Simulated Price (USD/barrel)")
plt.show()
```

```python
# --- Jump Diffusion Model ---

def jump_diffusion_sim(S0, mu, sigma, lamb, mu_j, sigma_j, T=1, steps=252, n_sims=100):
    dt = T/steps
    prices = np.zeros((steps, n_sims))
    for sim in range(n_sims):
        prices[0, sim] = S0
        for t in range(1, steps):
            N = np.random.poisson(lamb * dt)
            J = np.sum(np.random.normal(mu_j, sigma_j, N))
            drift = (mu - 0.5 * sigma**2) * dt
            diffusion = sigma * np.sqrt(dt) * np.random.normal()
            prices[t, sim] = prices[t-1, sim] * np.exp(drift + diffusion + J)
    return prices

mu = df["CRUDE_OIL"].pct_change().mean()
sigma = df["CRUDE_OIL"].pct_change().std()
S0 = df["CRUDE_OIL"].iloc[-1]
jd_prices = jump_diffusion_sim(S0, mu, sigma, lamb=0.1, mu_j=-0.02, sigma_j=0.05)
plt.figure(figsize=(10,5))
plt.figure(figsize=(10,5))
plt.plot(jd_prices, alpha=0.1, color='red')
plt.title("Jump Diffusion Simulations for Crude Oil Prices")
plt.xlabel("Days")
plt.ylabel("Price (USD/barrel)")
plt.legend()
plt.show()
```

```python
# --- NLTK Sentiment Analysis ---
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download required NLTK data
nltk.download('vader_lexicon', quiet=True)

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

headlines = [
    "Oil prices surge on geopolitical tensions",
    "Concerns over declining crude oil demand from slower economy",
    "OPEC cuts production supporting oil prices",
]

def get_nltk_sentiment(texts):
    sentiments = []
    for text in texts:
        scores = sia.polarity_scores(text)
        compound = scores['compound']
        
        # Classify based on compound score
        if compound >= 0.05:
            sentiment = "Positive"
        elif compound <= -0.05:
            sentiment = "Negative"
        else:
            sentiment = "Neutral"
        sentiments.append(sentiment)
    return sentiments

sentiments = get_nltk_sentiment(headlines)
for h, s in zip(headlines, sentiments):
    print(f"{h}  {s}")
```
